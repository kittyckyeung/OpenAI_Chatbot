# Ask Questions to ChatGPT in Jupyter

This notebook allows you to interact with OpenAI's ChatGPT directly from Jupyter. You can ask questions and receive answers using the OpenAI API.

```python
import os
import requests
import json
```

```python
key_location='C:\\Users\\user\\Desktop\\Coding\\OpenAPI_Key.txt'

with open(key_location, 'r') as file:
    key=file.readline().strip()
```

```python
def make_get_request(url):
    headers = {
        'Authorization': f'Bearer {key}'
    }
    response = requests.get(url, headers=headers, timeout=20)
    if response.status_code != 200:
        print('Request failed with status code:', response.status_code)
        return None
    result=response.json()
    return result
```

```python
def make_delete_request(url):
    headers = {
        'Authorization': f'Bearer {key}'
    }
    response = requests.delete(url, headers=headers, timeout=20)
    if response.status_code != 200:
        print('Request failed with status code:', response.status_code)
        return None
    result=response.json()
    return result
```

```python
def make_post_request(url, files, data):
    headers = {
        'Authorization': f'Bearer {key}'
    }
    response = requests.post(url, headers=headers, files=files, data=data, timeout=20)
    if response.status_code != 200:
        print('Request failed with status code:', response.status_code)
        return None
    result=response.json()
    return result
```

```python
list_url='https://api.openai.com/v1/files'
list_result=make_get_request(list_url)

if list_result:
    print(json.dumps(list_result, indent=2))
```

```python
upload_url='https://api.openai.com/v1/files'
files={
    "file": open("./data/data.jsonl", "rb")
}
data={
    "purpose": "fine-tune"
}    
upload_result=make_post_request(upload_url, files, data)

if upload_result:
    print(json.dumps(upload_result, indent=2))
upload_file_id=upload_result["id"]
```

```python
retrieve_url=f'https://api.openai.com/v1/files/{upload_file_id}'
retrieve_result=make_get_request(retrieve_url)

if retrieve_result:
    print(json.dumps(retrieve_result, indent=2))
```

```python
delete_url=f'https://api.openai.com/v1/files/{upload_file_id}'
delete_result=make_delete_request(delete_url)

if delete_result:
    print(json.dumps(delete_result, indent=2))
```
