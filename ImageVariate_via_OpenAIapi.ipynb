# Ask Questions to ChatGPT in Jupyter

This notebook allows you to interact with OpenAI's ChatGPT directly from Jupyter. You can ask questions and receive answers using the OpenAI API.

```python
import os
import requests
import json
```

```python
key_location='C:\\Users\\user\\Desktop\\Coding\\OpenAPI_Key.txt'

with open(key_location, 'r') as file:
    key=file.readline().strip()
```

```python
# Use the OpenAI API
url='https://api.openai.com/v1/images/variations'

files={
    "image": open("./images/otter2.png", "rb"),
}

data={
    "n": 1,
    "size": "1024x1024"
}
```

```python
def make_openai_api_request(url, files=None, data=None):
    headers = {
        'Authorization': f'Bearer {key}'
    }
    response = requests.post(url, headers=headers, files=files, data=data, timeout=20)
    if response.status_code != 200:
        print('Request failed with status code:', response.status_code)
    result=json.dumps(response.json(), indent=2)
    print(result)
```

```python
request_result=make_openai_api_request(url, files, data)

if request_result:
    print(json.dumps(request_result, indent=2))
```
