# Ask ChatGPT via Python (Translation Example)

This notebook demonstrates how to call the OpenAI API from Python to generate text and translate it using ChatGPT.

```python
from openai import OpenAI
import json
import requests
import os
```

```python
key_location='C:\\Users\\user\\Desktop\\Coding\\OpenAPI_Key.txt'

with open(key_location, 'r') as file:
    key=file.readline().strip()

client=OpenAI(api_key=key)
```

```python
def make_openai_request(prompt):
    response=client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return response.choices[0].message.content
```

```python
input_text="""
    Ben is a boy from Canada. His girlfriend's name is Alice. They love hiking together. He is working as hard as possible to earn more money to get marry with Alice, to give her a good home.
"""
```

```python
prompt=f"""
    Parsing the grammatical components from the following text.
    {input_text}
"""
```

```python
keywords=make_openai_request(prompt)
print(keywords)
```
