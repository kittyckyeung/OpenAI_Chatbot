# Ask Questions to ChatGPT in Jupyter

This notebook allows you to interact with OpenAI's ChatGPT directly from Jupyter. You can ask questions and receive answers using the OpenAI API.

```python
import os
import requests
import json
```

```python
key_location='C:\\Users\\user\\Desktop\\Coding\\OpenAPI_Key.txt'

with open(key_location, 'r') as file:
    key=file.readline().strip()
```

```python
# Use the OpenAI API
url='https://api.openai.com/v1/chat/completions'
```

```python
def make_openai_api_request(prompt):
    url = 'https://api.openai.com/v1/chat/completions'
    headers = {
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {key}'
    }
    data = {
        'model': 'gpt-3.5-turbo',
        'messages': [{'role': 'user', 'content': prompt}],
        'temperature': 0.7
    }
    response = requests.post(url, headers=headers, json=data, timeout=20)
    if response.status_code != 200:
        print('Request failed with status code:', response.status_code)
    result=json.dumps(response.json(), indent=2)
    print(result)
```

```python
prompt="Write a story of a traveling computer scientist in 2 paragraphs."
```

```python
make_openai_api_request(prompt)
```

```python
prompt="Translate the story into Spanish"
```

```python
make_openai_api_request(prompt)
```
